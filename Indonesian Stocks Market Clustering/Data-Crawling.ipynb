```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
path ='/content/drive/MyDrive/Colab Notebooks/IDX CLustering/daftar_saham.csv'
data = pd.read_csv(path)
```

```python
df = data.copy()
df.head()
```

```python
df.sort_values("ListingDate", ascending=False).head()
```

```python
df = df[df['ListingDate'] < '2017-01-01']
```

```python
df.set_index('Code', inplace=True)
```

```python
df.loc[["SUGI","TRIL","GOLL","KBRI","BTEL","CMPP","HDTX","LCGP","SCBD"]]
```

```python
df.drop(df.loc[["SUGI","TRIL","GOLL","KBRI","BTEL","CMPP","HDTX","LCGP","SCBD","APOL", "NIPS","GREN","TRIO"]].index, inplace=True)
```

```python
df.loc[["SUGI","TRIL","GOLL","KBRI","BTEL","CMPP","HDTX","LCGP","SCBD"]]
```

```python
df = df.reset_index()
```

```python
ticker_list = df['Code'].to_list()
str = ".JK"
ticker_list = [s + str for s in ticker_list]
ticker_list
company_list = df['Name'].to_list()
company_dict = dict(zip(company_list, ticker_list))
```

```python
ticker_list
```

```python
# download dataframe
df_new = yf.download(list(company_dict.values()), start="2016-12-31", end="2022-01-01")
```

```python
df_new
```

```python
df_new.reset_index()
```

```python
stocks = df_new.T
stocks
```

```python
stocks.reset_index(inplace=True)
stocks
```

```python
stocks = stocks.rename(columns={'level_0':'Attributes','level_1':'Symbols'})
df_2 = stocks.set_index(['Symbols', 'Attributes']).rename_axis(['Year'], axis=1).stack().unstack('Attributes').reset_index()
# df_2 = stocks.set_index(['Symbols', 'Attributes']).rename_axis(['Year'], axis=1).stack()
df_2
```
